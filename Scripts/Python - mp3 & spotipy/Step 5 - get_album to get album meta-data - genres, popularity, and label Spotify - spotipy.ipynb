# get album genre, popularity, label with album uri

```python
master_albums.head()
```

```python
album_dict_list = []
for i in np.arange(0, len(master_albums.uri_sp)):
    print(i)
    album_uri = master_albums.uri_sp[i]
    album_results = sp.album(album_uri)
    album_dict = {'uri_sp': album_uri,
                  'album_name': album_results['name'],
                  'genres_sp': album_results['genres'],
                  'popularity_sp': album_results['popularity'],
                  'label_sp': album_results['label']
        }
    album_dict_list.append(album_dict)
    time.sleep(np.random.uniform(2, 5))
```

```python
album_meta_data_1 = pd.DataFrame(album_dict_list)
album_meta_data_1.shape
```

```python
album_meta_data_1.head()
```

```python
album_meta_data_1.genres_sp.value_counts()
```
