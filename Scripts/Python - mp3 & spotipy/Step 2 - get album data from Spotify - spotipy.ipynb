### first attempt - died b/c of null in artist name. fixed above now

```python
album_search_list = []
not_found_list = []
for i in np.arange(0, len(tag_data_no_dupes.album_clean)):
    print(i)
    album_to_search = tag_data_no_dupes['album_clean'][i]
    artist_to_search = tag_data_no_dupes['artist_clean'][i]
    search_results = sp.search(q = "album:" + album_to_search + ' ' "artist:" + artist_to_search, type = "album")
    if len(search_results['albums']['items']) == 0:
        not_found_dict = {
            'album_to_search': album_to_search,
            'artist_to_search': artist_to_search
        }
        not_found_list.append(not_found_dict)
    else:
        album_dict = {
        'album_to_search': album_to_search,
        'artist_to_search': artist_to_search,
        'artist_name_sp': search_results['albums']['items'][0]['artists'][0]['name'],
        'artist_uri_sp': search_results['albums']['items'][0]['artists'][0]['uri'],
        'album_name_sp': search_results['albums']['items'][0]['name'],
        'uri_sp': search_results['albums']['items'][0]['uri'], 
        'total_tracks_sp': search_results['albums']['items'][0]['total_tracks'],
        'release_date_sp': search_results['albums']['items'][0]['release_date'],
        'release_type_sp': search_results['albums']['items'][0]['type'],
        }
        album_search_list.append(album_dict)
    time.sleep(np.random.uniform(2, 5))
```
