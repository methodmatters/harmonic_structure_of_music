# search for tracks with ids

```python
master_track_list = []
for i in np.arange(0, len(master_albums.uri_sp)):
    print(i)
    uri_loop = master_albums.uri_sp[i]
    # print(uri_loop)
    track_dict = sp.album_tracks(uri_loop)
    if track_dict['total'] > 50:
        num_tracks = 50
    else:
        num_tracks = track_dict['total']
    for j in np.arange(0,num_tracks):
        track_number = track_dict['items'][j]['track_number']
        explicit = track_dict['items'][j]['explicit']
        track_name = track_dict['items'][j]['name']
        track_type = track_dict['items'][j]['type']
        track_duration_ms = track_dict['items'][j]['duration_ms']
        track_disc_number = track_dict['items'][j]['disc_number']
        track_uri = track_dict['items'][j]['uri']
        master_track_dict = {'album_uri': uri_loop,
                             'num_tracks': num_tracks,
                             'track_number': track_number,
                             'explicit': explicit,
                             'track_name': track_name,
                             'track_type': track_type,
                             'track_duration_ms': track_duration_ms,
                             'track_disc_number': track_disc_number,
                             'track_uri': track_uri
        }
        master_track_list.append(master_track_dict)
    time.sleep(np.random.uniform(2, 5))
```

```python
# get stuff for zappa album
zappa_id = 'spotify:album:51cjNmDteqloO1E3O0wrfd'
track_dict_zappa = sp.album_tracks(zappa_id, offset = 50)
num_tracks_zappa = track_dict_zappa['total']
num_tracks_zappa
```

```python
master_track_list_zappa = []
for j in np.arange(0,len(track_dict_zappa['items'])):
    # print(j)
    track_number = track_dict_zappa['items'][j]['track_number']
    explicit = track_dict_zappa['items'][j]['explicit']
    track_name = track_dict_zappa['items'][j]['name']
    track_type = track_dict_zappa['items'][j]['type']
    track_duration_ms = track_dict_zappa['items'][j]['duration_ms']
    track_disc_number = track_dict_zappa['items'][j]['disc_number']
    track_uri = track_dict_zappa['items'][j]['uri']
    master_track_dict = {'album_uri': zappa_id,
                         'num_tracks': track_dict_zappa['total'],
                         'track_number': track_number,
                         'explicit': explicit,
                         'track_name': track_name,
                         'track_type': track_type,
                         'track_duration_ms': track_duration_ms,
                         'track_disc_number': track_disc_number,
                         'track_uri': track_uri
    }
    master_track_list_zappa.append(master_track_dict)
```

```python
missing_zappa_tracks = pd.DataFrame(master_track_list_zappa)
```
