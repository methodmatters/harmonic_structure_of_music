```python
# go through directories and look for all mp3 files
root = "D:\\Music\\"
path = "D:\\Music\\"
pattern = "*.mp3"
from fnmatch import fnmatch
all_mp3_files = []
for path, subdirs, files in os.walk(root):
    for name in files:
        if fnmatch(name, pattern):
            all_mp3_files.append(os.path.join(path, name))
```

```python
%%time
song_info_list =[]
for i in np.arange(0, len(selected_files)):
    song = selected_files[i]
    try:
        mp3file = MP3(song, ID3=EasyID3)
        mp3file_dict = dict(mp3file)
        mp3file_dict['album_path'] = [song]
        song_info_list.append(pd.DataFrame().from_dict(mp3file_dict, orient='index').T)
    except Exception as err:
        print(song)
        print(err)

```

```python
# remove all punctuation and  replace with dash
import re
import string
# stripping accents
import unicodedata
regex = re.compile('[%s]' % re.escape(string.punctuation))
def remove_punct(s):  # From Vinko's solution, with fix.
    return regex.sub('', s)


# define our text cleaning function
def clean_artist_names(raw_artist):
    # remove the end part of the artist name
    root_artist = re.split('&|/|,|feat.|Feat.|Disc|disc', raw_artist)[0]
    lower_case = root_artist.lower()
    no_punct = remove_punct(lower_case)
    no_whitespace = ' '.join(no_punct.split())
    # remove accented characters - replace them with non-accented equivalents
    no_accents = unicodedata.normalize('NFKD', no_whitespace).encode('ASCII', 'ignore').decode()

    return(no_accents)  

```

```python
tag_data_clean['artist_clean'] = [clean_artist_names(x) for x in tag_data_clean['master_artist']]
[x for x in np.unique(tag_data_clean['artist_clean'])]
```

```python
tag_data_clean['album_clean'] =  [clean_artist_names(x) for x in tag_data_clean['album']]
[x for x in np.unique(tag_data_clean['album_clean'])]
```
